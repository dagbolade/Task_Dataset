Data Cleaning

```python
# Check for missing values
print("Missing values:\n", df_Sales.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer

#create an instance of the SimpleImputer class
imputer = SimpleImputer(strategy='most_frequent')

# Replace missing values with the imputed values
df_imputed = pd.DataFrame(imputer.fit_transform(df_Sales), columns=df_Sales.columns)

# Check the imputed data
print("Imputed Data:")
print(df_imputed.head())
```

```python
# Check for missing values
print("Missing values:\n", df_imputed.isnull().sum())
```

Data Exploration from the Sales Data

```python
#calculate the total revenue
total_revenue = df_imputed['Revenue £'].sum()
formatted_revenue = '{:,.0f}'.format(total_revenue)

print('Total Revenue: £', formatted_revenue)
```

Visualisation: Revenue Trend Over Time

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import StrMethodFormatter

# Convert 'Document Date' to datetime type
df_imputed['Document Date'] = pd.to_datetime(df_imputed['Document Date'])

# Calculate monthly revenue
monthly_revenue = df_imputed.groupby(df_imputed['Document Date'].dt.to_period('M'))['Revenue £'].sum()

# Bar plot of revenue over time
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_revenue.index.astype(str), y=monthly_revenue.values)
plt.title('Revenue Trend Over Time')
plt.xlabel('Month')
plt.ylabel('Revenue (£)')
plt.xticks(rotation=45)
# Format y-axis tick labels
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()
```

Top Selling Items

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import StrMethodFormatter

# Merge df_imputed with df_items to get item names
df_merged = pd.merge(df_imputed, df_items[['ItemCode', 'ItemName']], on='ItemCode', how='left')

# Convert 'Quantity' column to numeric type
df_merged['Quantity'] = pd.to_numeric(df_merged['Quantity'], errors='coerce')

# Group by month and item name, calculate total quantity sold
monthly_item_sales = df_merged.groupby([df_merged['Document Date'].dt.to_period('M'), 'ItemName'])['Quantity'].sum()

# Get top-selling items for each month
top_items_by_month = monthly_item_sales.groupby(level=0).nlargest(3).reset_index(level=0, drop=True).reset_index()

# Plot top-selling items for each month
plt.figure(figsize=(12, 8))
sns.barplot(x='Document Date', y='Quantity', hue='ItemName', data=top_items_by_month)
plt.title('Top-Selling Items per Month')
plt.xlabel('Month')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.legend(title='Item Name', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

Compare the performance of the Account Managers

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Merge df_imputed with df_Am to get account manager names
df_merged = df_imputed.merge(df_Am, left_on='Account Manager', right_on='Account Manager code', how='left')

# Group by account manager name and calculate total sales and revenue
account_manager_performance = df_merged.groupby('Account Manager name').agg({'Quantity': 'sum', 'Revenue £': 'sum'}).reset_index()

# Sort by total sales in descending order
account_manager_performance = account_manager_performance.sort_values('Quantity', ascending=False)

# Bar plot of sales by account manager
plt.figure(figsize=(10, 6))
sns.barplot(x='Quantity', y='Account Manager name', data=account_manager_performance)
plt.title('Sales Performance by Account Manager')
plt.xlabel('Quantity')
plt.ylabel('Account Manager')
# Format x-axis tick labels
plt.gca().xaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()

# Sort by total revenue in descending order
account_manager_performance = account_manager_performance.sort_values('Revenue £', ascending=False)

# Bar plot of revenue by account manager
plt.figure(figsize=(10, 6))
sns.barplot(x='Revenue £', y='Account Manager name', data=account_manager_performance)
plt.title('Revenue Performance by Account Manager')
plt.xlabel('Revenue (£)')
plt.ylabel('Account Manager')
# Format x-axis tick labels
plt.gca().xaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()
```

Visualisation: Revenue by Customer

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Merge df_imputed with df_Customers to get customer names
df_merged = df_imputed.merge(df_Customers, on='Customer ID', how='left')

# Replace non-numeric values in 'Revenue £' column with NaN
df_merged['Revenue £'] = pd.to_numeric(df_merged['Revenue £'], errors='coerce')

# Group by customer name and calculate total revenue and returns
customer_performance = df_merged.groupby('CustomerName').agg({'Revenue £': 'sum', 'Cancelled': 'sum'}).reset_index()

# Sort by total revenue in descending order
customer_performance = customer_performance.sort_values('Revenue £', ascending=False)

# Select top 15 customers by revenue
top_revenue_customers = customer_performance.head(15)

# Bar plot of revenue by customer
plt.figure(figsize=(10, 6))
sns.barplot(x='Revenue £', y='CustomerName', data=top_revenue_customers, order=top_revenue_customers['CustomerName'])
plt.title('Top 15 Customers by Revenue')
plt.xlabel('Revenue (£)')
plt.ylabel('Customer Name')
# Format x-axis tick labels
plt.gca().xaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()

```

sales by region

```python
# Merge df_imputed with df_Customers to get customer regions
df_merged = df_imputed.merge(df_Customers[['Customer ID', 'Customer Address Region']], on='Customer ID', how='left')

# Group by region and calculate total revenue
region_sales = df_merged.groupby('Customer Address Region')['Revenue £'].sum().reset_index()

# Sort regions by total revenue in descending order
region_sales = region_sales.sort_values('Revenue £', ascending=False)

# Select top 20 regions by revenue
top_regions = region_sales.head(20)

# Bar plot of sales by region
plt.figure(figsize=(10, 6))
sns.barplot(x='Revenue £', y='Customer Address Region', data=top_regions, order=top_regions['Customer Address Region'])
plt.title('Top 20 Regions by Sales')
plt.xlabel('Revenue (£)')
plt.ylabel('Region')
# Format x-axis tick labels
plt.gca().xaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()
```
